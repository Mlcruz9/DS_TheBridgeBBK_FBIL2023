```python
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
```

```python
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = bs(response.text, 'lxml')
```

```python
print(response)
html = response.content
soup = bs(html, "lxml")
```
